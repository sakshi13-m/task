Write a Python script that finds the top 10 colleges for Civil Engineering Branch based on cutoff
ranks in GOPENS/GOPENH categories for Savitribai Phule Pune University?

Firstly importing pandas for data manipulation and to extract required data from given pdf i used PyPDF2 and regex which help me sort the required data


```
import pandas as pd
```

**Here,I created a loop to get data only for the required university and branch.**

```
import PyPDF2  
import re
pf= open('college-cutoff.pdf', 'rb')  
    
# creating a pdf reader object  
Reader = PyPDF2.PdfFileReader(pf)  
    
# printing number of pages in pdf file  
#print(Reader.numPages)  
    
i=0
c=0
while(i<1303):
  pageObj = Reader.getPage(i)  
  text= pageObj.extractText()
  if(re.search('Home University : Savitribai Phule Pune University',text) and re.search('Civil Engineering',text)):
    print(i+1)
    print(text)
    print() 
    c+=1
  i+=1
print(c)
    
# closing the pdf file object   
```

Using the above data i manually created a excel file with the name of college its code and cutoff rank for GOPENS/GOPENH which have 79 entries in it this file will ease out my task.

Next created a dataframe using pandas and reading the file in csv format with required columns and rows

```
df=pd.read_csv('college-cutoff.csv',usecols=['College name','clg code','GOPENS/GOPENH','subject code'],encoding='UTF-8')
df=df[:79]
df
```

As the data type of few columns are changed due to csv conversion so i changed it back to original ones.

```
df['clg code']=df['clg code'].astype('int64')
df['GOPENS/GOPENH']=df['GOPENS/GOPENH'].astype('int64')
df['subject code']=df['subject code'].astype('int64')
df
```

Finally,using sort_values function by cutoff rank will give us the top 10 college as required in the task

```
gd=df.sort_values(by=['GOPENS/GOPENH'])
gd.head(10)
```
